### **Outlier Detection**
identify potential outlier in numerical variables.

```python

def plot_numerical_outliers(df, features, cols_per_row=3):
    num_features = len(features)
    
    # NEW LOGIC: Calculate rows without the math module
    num_rows = num_features // cols_per_row
    if num_features % cols_per_row != 0:
        num_rows += 1
        
    fig, axes = plt.subplots(num_rows, cols_per_row, figsize=(10,12))
    axes = axes.flatten()
    
    for i, col in enumerate(features):
        sns.boxplot(y=df[col], ax=axes[i])
        axes[i].set_title(f'Distribution of {col}')
        axes[i].set_ylabel('') 
        
        
    plt.tight_layout()
    plt.show()

# Execute the visualization
plot_numerical_outliers(df, num_cols, cols_per_row=3)
```

### **Quantifying outliers**
boxplots tells us outliers exist, doesn't tell how many or are they worth keeping or removing

```python
outlier_summaary = []
for col in num_cols :
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 -Q1
    lower = Q1 - 1.5*IQR
    upper = Q3 + 1.5*IQR
    n_outliers = ((df[col] < lower) | (df[col] > upper)).sum()
    pct_outliers = round(n_outliers / len(df) *100,2)
    outlier_summaary.append({"feature": col,
                             "n_outlers":n_outliers,
                             "pct_outliers" :pct_outliers})
outler_df = pd.DataFrame(outlier_summaary).sort_values("pct_outliers",ascending=False)
outler_df
```

### **Inference**
- some numerical columns exhibit outliers,but the affected % of rows is small for most features.
- `MonthlyIncome`,`YearsAtCompany`,`TotalWorkingYears` have high values
- Since they may represent genuine employee characteristics rather than data error, we should keep them
